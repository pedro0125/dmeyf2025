Creacion Target

Si foto_mes >= 202106 → target = NULL
Si foto_mes = 202105 → target = 'BAJA+1' (si no aparece en 202106)
Si foto_mes < 202105:
    - Si no aparece en el mes siguiente → 'BAJA+1'
    - Si aparece en el mes siguiente pero no en el segundo → 'BAJA+2'
    - Si aparece en los dos meses siguientes → 'CONTINUA'


```python
%%sql
COPY (
    with base as (
        SELECT 
            c.*,
            lead(foto_mes, 1) OVER (PARTITION BY numero_de_cliente ORDER BY foto_mes) as foto_mes_1,
            lead(foto_mes, 2) OVER (PARTITION BY numero_de_cliente ORDER BY foto_mes) as foto_mes_2
        FROM competencia_01_crudo c
    ),
    etiquetado as (
        SELECT
            base.* exclude (foto_mes_1, foto_mes_2),
            case
                when foto_mes < 202106 then
                    case
                        when foto_mes < 202105 then
                            case
                                when foto_mes_1 is null then 'BAJA+1'
                                when foto_mes_2 is null then 'BAJA+2'
                                else 'CONTINUA'
                            end
                        else case when foto_mes_1 is null then 'BAJA+1' end
                    end
            end as clase_ternaria
        FROM base
    )
    select * from etiquetado
) TO "{{dataset_path + dataset_target_file}}" (HEADER, DELIMITER ',');
```

```python
%%sql
create or replace table competencia_01 as
select
    *
from read_csv_auto("{{dataset_path + dataset_target_file}}")
```

```python
%%sql
select
    foto_mes
    , count(*) as cantidad -- cuenta cuantos casos hay en cada foto_mes
                           -- y lo guarda en un campo llamado cantidad
from competencia_01_crudo
group by foto_mes
```
